# range #

```python
for i in range(1,10,2):
    print(i)
```

# else loops #

```python
i=1
for i in range(10):
    print(i)
else:
    print('outside loop')
```

# Match Case #

```python
vote=int(input("please enter your vote"))
bjp=100
sp=50
bsp=34
congress=80
match vote:
    case 1:
        print('you voted for bjp')
        bjp+=1
        print(bjp)
    case 2:
        print('you voted for sp')
        sp+=1
    case 3:
        print('you voted for bsp')
        bsp+=1
    case 4:
        print('you voted for congress')
        congress+=1
    case _:
        print('invalid your vote')
```
