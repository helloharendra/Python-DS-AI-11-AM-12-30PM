### BITWISE OPERATORS

```python
# Binary Example
a = 5  # → 0101
b = 3   #→ 0011
```

```python
#AND
#1 only if both bits are 1

a = 5
b = 3
print(a & b)   # Output: 1
```

# OR (|)
print(a | b)   # Output: 7

```python
print(a | b)   # Output: 7
```

```python
3. XOR (^)

👉 1 if bits are different

print(a ^ b)   # Output: 6
```

```python
#if statement
age = 18
if age >= 18:
    print("You can vote")
```

```python
NOT (~)

👉 Inverts bits

print(~5)   # Output: -6
```

```python
Left Shift (<<)

👉 Shift bits left (multiply by 2)

print(5 << 1)   # Output: 10
```

```python
Right Shift (>>)

👉 Shift bits right (divide by 2)

print(5 >> 1)   # Output: 2
0101 → 0010
```

```python
Example
num = int(input("Enter number: "))

if num & 1:
    print("Odd")
else:
    print("Even")
```

```python
# if-else
age = 16
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# if-elif-else
marks = 75

if marks >= 90:
    print("A Grade")
elif marks >= 70:
    print("B Grade")
else:
    print("C Grade")
```

```python
# for loop
for i in range(5):
    print(i)
```

```python
# while loop
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# break (stop loop)
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
# continue (skip iteration)
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# pass (do nothing)
for i in range(5):
    pass
```

```python
# Nested Control Flow
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# Match Case (Python 3.10+)
day = 2

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case _:
        print("Other Day")
```

```python
IF, ELIF, ELSE
age = int(input("Enter age: "))

if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teenager")
else:
    print("Child")
🔁 2. FOR LOOP
for i in range(1, 6):
    print(i)

Output:

1 2 3 4 5
🔁 3. WHILE LOOP
i = 1

while i <= 5:
    print(i)
    i += 1
⛔ 4. BREAK
for i in range(10):
    if i == 5:
        break
    print(i)

👉 Stops loop at 5

⏭️ 5. CONTINUE
for i in range(5):
    if i == 2:
        continue
    print(i)

👉 Skips 2

🔄 6. PASS
for i in range(5):
    pass   # placeholder
🚀 COMBINED PROGRAM (IMPORTANT)

👉 Uses bitwise + control flow together

# Print even and odd numbers separately using bitwise

even = []
odd = []

for i in range(1, 21):
    if i & 1:
        odd.append(i)
    else:
        even.append(i)

print("Even Numbers:", even)
print("Odd Numbers:", odd)
🧪 ADVANCED PROGRAM (MINI PROJECT)

👉 Count number of 1s in binary

num = int(input("Enter number: "))
count = 0

while num > 0:
    count += num & 1
    num = num >> 1

print("Number of 1 bits:", count)
```
