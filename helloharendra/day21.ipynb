# Sum of n natural number

```python
n = 10
total = 0

for i in range(1, n + 1):
    total += i

print(total)
```

```python
n = int(input('Enter the number fo find sum of natural num'))
total = 0

for i in range(1, n + 1):
    total += i

print(total)
```

```python
l2=[10,10,20,40,30,40,80,80,50,60,100,100,44,44]
```

```python
for i in range(len(l2)):
    if l2[i] == 80:
        print(i)
```

```python
duplicates=set()
seen=set()

for item in l2:
    if item in seen:
        duplicates.add(item)
    else:
        seen.add(item)
print(list(duplicates))
```
