# While Loops

- Runs while condition is True
- The while loop runs as long as a given condition is True.

```python
i=0

while i<=3:
    print(i)
    i+=1
```

```python
i=10

while i>=1:
    print(i)
    i-=1
```

```python
signal=input('Enter your signal color Like. red, green, yellow')

while signal=='red':
    print('stop')
    break
print('go')
```
