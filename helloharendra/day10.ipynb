# Ternary Operator

- Short form of if-else
- The ternary operator in Python is a one-line shortcut for writing simple if-else conditions.


```python

age=18
if age>=18:print('can vote')
else:
    print("can't vote  ")
```

```python
result=print("can vote") if age>=18 else print("can't vote")
```

```python
result='adult' if age>=18 else 'minor'
```

```python
print(result)
```

# Loops

- for Loop 

- The for loop is used to iterate over a sequence like a list, tuple, string, or range.

```python
for b in range(5):
    print(b)
```

```python
for i in range(1000):
    print(f'{i}radhe')
```
