# AND
#### 1 only if both bits are 1 



```python
a = 5
b = 3
print(a & b)   # Output: 1
```

# OR (|)
#### 1  if any of  both bits are 1 


```python
print(a | b)   # Output: 7
```

# XOR (^)
- if both bit is true return 0 else 1

```python


print(a ^ b)   # Output: 6
```

```python
#if statement
age = 18
if age >= 18:
    print("You can vote")
```

```python


print(~5)   # Output: -6
```

```python




print(5 << 1)   # Output: 10
```

# Right Shift (>>)

### 👉 Shift bits right (divide by 2)


```python

print(5 >> 1)   # Output: 2
```

# Example

```python

num = int(input("Enter number: "))

if num & 1:
    print("Odd")
else:
    print("Even")
```

```python
# if-else
age = 16
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# if-elif-else
marks = 75

if marks >= 90:
    print("A Grade")
elif marks >= 70:
    print("B Grade")
else:
    print("C Grade")
```

```python
# for loop
for i in range(5):
    print(i)
```

```python
# while loop
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# break (stop loop)
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
# continue (skip iteration)
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# pass (do nothing)
for i in range(5):
    pass
```

```python
# Nested Control Flow
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# Match Case (Python 3.10+)
day = 2

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case _:
        print("Other Day")
```
