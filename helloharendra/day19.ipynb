### DSA (Introduction to Data Structures & Algorithms)

Data Structures are ways to store and organize data so that it can be used efficiently.
Algorithms are step-by-step procedures to solve a problem.

#### Why DSA is important:

- Improves problem-solving skills
- Helps write optimized code
- Used in software development, AI, and interviews.

#### Time and Space Complexity

Time Complexity measures how fast an algorithm runs.
Space Complexity measures how much memory it uses.

#### Types:
- Constant Time → O(1)
- Linear Time → O(n)
- Logarithmic Time → O(log n)
- Quadratic Time → O(n²)
#### Concepts:
- Best Case
- Worst Case
- Average Case


### Arrays (Lists in Python)
Array is a collection of elements stored in contiguous memory. In low-level languages like C, elements are stored one after another in memory, which allows fast access using indexes. In Python, we use lists, which behave like dynamic arrays.



## Operations on Arrays
#### Traversal
Visiting each element one by one.


```python
arr=[1,2,3,4,5]

for i in range(len(arr)):
    print(arr[i])
```

```python
l1=[1,2,3,4,5]

for i in range(len(l1)):
    print(l1[i])
```

```python
l2=[1,2,3,4,5]

l3=len(l2)

for i in range(l3):
    print(l2[i])
```

```python
print(l2[0])
```
