# else with Loops
- Runs when loop completes normally (not stopped by break)

```python
for i in range(5):
    print(i)
print('ouside loop')
```

```python
for i in range(5):
    print(i)
else:
    print('ouside loop')
```

# match-case
- Alternative of if else

```python
day=2

match day:
    case 1:
        print('sunday')
    case 2:
        print('monday')
    case 3:
        print('tuesday')
    case 4:
        print('wednesday')
    case 5:
        print('thursday')
    case 6:
        print('friday')
    case 7:
        print('saturday')
    case _:
        print('invalid day')
    
```

```python
bjp = 0
sp = 50
bsp = 60
inc = 40

print('press 1 for BJP')
print('press 2 for SP')
print('press 3 for BSP')
print('press 4 for INC')

vote = int(input('Please press the desired button: '))

match vote:
    case 1:
        print('You voted for BJP')
        bjp += 1
    case 2:
        print('You voted for SP')
        sp += 1
    case 3:
        print('You voted for BSP')
        bsp += 1
    case 4:
        print('You voted for Congress')
        inc += 1
    case _:
        print('Invalid vote')

# Final Result
print("\n--- Updated Vote Count ---")
print("BJP:", bjp)
print("SP:", sp)

print("BSP:", bsp)
print("INC:", inc)
```

```python
bjp = 0
sp = 50
bsp = 60
inc = 40

while True:
    print('\npress 1 for BJP')
    print('press 2 for SP')
    print('press 3 for BSP')
    print('press 4 for INC')
    print('press 0 to exit')

    vote = int(input('Please press the desired button: '))

    match vote:
        case 1:
            print('You voted for BJP')
            bjp += 1
        case 2:
            print('You voted for SP')
            sp += 1
        case 3:
            print('You voted for BSP')
            bsp += 1
        case 4:
            print('You voted for Congress')
            inc += 1
        case 0:
            print('Voting ended')
            break
        case _:
            print('Invalid vote')

# Final Result
print("\n--- Final Vote Count ---")
print("BJP:", bjp)
print("SP:", sp)
print("BSP:", bsp)
print("INC:", inc)
```
