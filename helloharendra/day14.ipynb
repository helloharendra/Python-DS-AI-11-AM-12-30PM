# List Comprehension

- List comprehension is a shortcut to create lists in one line instead of using loops.


```python
list1=[1,2,3,4,5,6]
```

```python
print(list1)
```

```python
print(list1[3])
```

```python

list2 = [x for x in range(5)]
```

```python
print(list2)
```

```python
squares=[x**2 for x in range(10)]
```

```python
print(squares)
```

```python
cubes=[x**3 for x in range(10)]
print(cubes)
```

```python
even_numbers = [x for x in range(100) if x % 2 == 0]
```

```python
print(even_numbers)
```

```python
num=[12,10,7,8,9,45,36,77]

result = ["Even" if x % 2 == 0 else "odd" for x in num]
print(result)
```

```python
name='naya prayog academy'

print(name.upper())
```

```python
name =['Prashant','vijay', 'roshan']

upper_name=[name.upper() for name in name]
print(upper_name)
```

```python
lower_name=[name.lower() for name in name]
print(lower_name)
```

```python
num=[12,10,7,8,9,45,36,77]

result = [print(x) if x >10 else print('') for x in num]
print(result)
```
