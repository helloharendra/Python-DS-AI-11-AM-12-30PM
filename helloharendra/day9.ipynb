# card number validation

```python
cardnum=input('Enter your 16 digit credit/debit card number')
if len(cardnum)==16:
    print('valid card')
    cardpin=input('Enter your card pin')
    if len(cardpin)==4:
        print('valid pin')
        cvv=input('enter your cvv')
        if len(cvv)==3:
            print('valid cvv')
            otp=input('Enter 4 digit otp')
            if len(otp)==4:
                print('Payment successful')
            else:
                print('incorrect otp Payment failed')
        else:
            print('invalid cvv')
    else:
        print('invalid pin')
else:
    print('invalid card number')
```
