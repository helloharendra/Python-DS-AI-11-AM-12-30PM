```python
num=int(input('Enter you number to check'))
if num>1:
    is_prime=True
    
    for i in range(2,num):
        if num % i == 0:
            is_prime=False
            break
    if is_prime:
            print('prime')
    else:
            print('not prime')
else:
    print('Not Prime')
```
