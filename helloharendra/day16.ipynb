# Exception Handling in Python 

Exception handling is used to handle runtime errors so your program doesn’t crash unexpectedly.

### Why we use it
- Prevents program from crashing
- Handles errors gracefully
- Improves user experience
- Helps debug issues easily

```python
try:
    num = int(input("Enter number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input please enter valid number")

except Exception as e:
    print('Exception',e)

finally:
    print('Task Done')
```

```python
x = 3
assert x > 0, "x must be positive"
```

```python
try:
    open('text.txt','r')

except Exception as e:
    print("Error ",e)

```

```python
with open("text.txt", "r") as f:
    data = f.read()
```

```python
def count_up(n):
    for i in range(n):
        yield i
for num in count_up(5):
    print(num)
```

```python
count_up(12)
```

# Recursion

- A function call itself


```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(50))
```

```python
for i in range(5):
    print("*")
```

```python
print(factorial(5))
```

```python
print(factorial(8))
```
