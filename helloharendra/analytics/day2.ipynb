```python
# ------------------------------
# 6. HANDLING MISSING DATA
# ------------------------------
print("\n" + "="*60)
print("6. MISSING DATA (NaN)")
print("="*60)
```

```python
# Create a DataFrame with NaN
df_nan = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})
print("With NaNs:\n", df_nan)
```

```python
print("\nIs NaN?\n", df_nan.isnull())
```

```python
# Detect missing

print("\nAny missing per column?\n", df_nan.isnull().any())
```

```python
# Drop rows with any NaN
df_clean1 = df_nan.dropna()
print("\nDrop rows with any NaN:\n", df_clean1)
```

```python
# Fill NaN with a value
df_filled = df_nan.fillna(0)
print("\nFill NaN with 0:\n", df_filled)
```

```python
# Fill with column mean
df_nan['A'] = df_nan['A'].fillna(df_nan['A'].mean())
print("\nAfter filling column A with its mean:\n", df_nan)
```

```python


# ------------------------------
# 7. OPERATIONS ON DATA
# ------------------------------
print("\n" + "="*60)
print("7. BASIC OPERATIONS")
print("="*60)

# Arithmetic
df['Bonus'] = df['Score'] * 0.1
print("Added Bonus column:\n", df)

# Apply function
df['Score_Squared'] = df['Score'].apply(lambda x: x**2)
print("\nScore squared using apply:\n", df[['Score', 'Score_Squared']])

# Vectorized operation
df['Age_Next_Year'] = df['Age'] + 1
print("\nAge next year:\n", df[['Age', 'Age_Next_Year']])

# ------------------------------
# 8. GROUP BY & AGGREGATION
# ------------------------------
print("\n" + "="*60)
print("8. GROUP BY")
print("="*60)

# Group by City and calculate mean Score
grouped = df.groupby('City')['Score'].mean()
print("Mean Score by City:\n", grouped)

# Multiple aggregations
agg_df = df.groupby('City').agg({
    'Score': ['mean', 'max', 'min'],
    'Age': 'mean'
})
print("\nMultiple aggregations:\n", agg_df)

# Group by and count
print("\nCount per City:\n", df.groupby('City').size())

# ------------------------------
# 9. MERGING & CONCATENATION
# ------------------------------
print("\n" + "="*60)
print("9. MERGE & CONCAT")
print("="*60)

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Score': [85, 92, 78]})

# Inner merge (like SQL JOIN)
merged = pd.merge(df1, df2, on='ID', how='inner')
print("Inner merge:\n", merged)

# Left merge
left_merged = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft merge:\n", left_merged)

# Concatenation (stacking)
df_a = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_b = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
concat_rows = pd.concat([df_a, df_b], axis=0)   # vertical stack
print("\nConcatenate rows:\n", concat_rows)

concat_cols = pd.concat([df_a, df_b], axis=1)   # horizontal
print("\nConcatenate columns:\n", concat_cols)

# ------------------------------
# 10. PIVOT TABLES & CROSS-TABS
# ------------------------------
print("\n" + "="*60)
print("10. PIVOT TABLES")
print("="*60)

# Create sample sales data
sales = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 130],
    'Region': ['East', 'West', 'East', 'West']
})

pivot = pd.pivot_table(sales, values='Sales', index='Product', columns='Region', aggfunc='sum', fill_value=0)
print("Pivot Table (Sales by Product and Region):\n", pivot)

# Cross-tab (frequency table)
crosstab = pd.crosstab(sales['Product'], sales['Region'])
print("\nCross-tab (counts):\n", crosstab)

# ------------------------------
# 11. APPLY & LAMBDA FUNCTIONS
# ------------------------------
print("\n" + "="*60)
print("11. APPLY FOR CUSTOM OPERATIONS")
print("="*60)

def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    else:
        return 'C'

df['Grade'] = df['Score'].apply(grade)
print("Grades added:\n", df[['Name', 'Score', 'Grade']])

# Apply to multiple columns at once
df['Sum_Cols'] = df[['Age', 'Score']].apply(lambda row: row['Age'] + row['Score'], axis=1)
print("\nSum of Age and Score:\n", df[['Age', 'Score', 'Sum_Cols']])

# ------------------------------
# 12. TIME SERIES (optional basic)
# ------------------------------
print("\n" + "="*60)
print("12. TIME SERIES (BASIC)")
print("="*60)

date_range = pd.date_range(start='2024-01-01', periods=5, freq='D')
ts_df = pd.DataFrame({'Value': [10, 20, 15, 30, 25]}, index=date_range)
print("Time series DataFrame:\n", ts_df)

# Resample to 2-day frequency
print("\nResampled (2D mean):\n", ts_df.resample('2D').mean())

# ------------------------------
# 13. BASIC PLOTTING (optional)
# ------------------------------
print("\n" + "="*60)
print("13. SIMPLE PLOT (will show a window)")
print("="=60)

# Uncomment to see plot
# df.plot(kind='bar', x='Name', y='Score', title='Scores by Student')
# plt.tight_layout()
# plt.show()

# ------------------------------
# 14. RESET INDEX & SORTING
# ------------------------------
print("\n" + "="*60)
print("14. SORTING & RESET INDEX")
print("="=60)

# Sort by Score descending
df_sorted = df.sort_values('Score', ascending=False)
print("Sorted by Score (desc):\n", df_sorted)

# Reset index after filtering
filtered = df[df['Score'] > 80].reset_index(drop=True)
print("\nFiltered and reset index:\n", filtered)

# ------------------------------
# 15. CLEAN UP (delete example files)
# ------------------------------
import os
os.remove('students.csv')   # clean up
print("\nCleaned up temporary files.")

print("\n" + "="*60)
print("PANDAS TUTORIAL COMPLETED SUCCESSFULLY!")
print("="*60)
How to use this with your students:
Save the code as pandas_tutorial.py

Run it with: python pandas_tutorial.py

They can modify each section to experiment

Explain each section as they run it

Suggested class flow (3-4 hours):
Time	Topic
0:00	Intro to Pandas & Series
0:20	DataFrames (creation, I/O)
0:40	Exploration (.info, .describe)
1:00	Indexing (loc, iloc, boolean)
1:30	Handling NaN (dropna, fillna)
2:00	GroupBy & Aggregation
2:30	Merge, Concat, Pivot
3:00	Apply functions
3:30	Time series & plotting
3:45	Q&A, exercises
Homework idea:
Ask them to load a real CSV (e.g., from Kaggle's Titanic dataset) and perform:

Clean missing values

Group by a column

Create a pivot table

Add a calculated column

Save the cleaned result

Would you like me to also provide exercise problems or a Jupyter notebook version for interactive teaching?


```
